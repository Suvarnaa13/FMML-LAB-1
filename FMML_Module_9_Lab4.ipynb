## Using learnt representations
In practice, very few people train an entire Convolutional Network from scratch (with random initialization), because it is relatively rare to have a dataset of sufficient size. Instead, it is common to pretrain a ConvNet on a very large dataset (e.g. ImageNet, which contains 1.2 million images with 1000 categories), and then use the ConvNet either as an initialization or a fixed feature extractor for the task of interest.


We'll train a model to classify ants and bees. We have about 120 training images each for ants and bees. There are 75 validation images for each class. Usually, this is a very small dataset to generalize upon, if trained from scratch. Since we are using transfer learning, we should be able to generalize reasonably well.

This dataset is a very small subset of imagenet.




```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, SubsetRandomSampler
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
import time
import os
```

```python
# Device configuration (whether to run on GPU or CPU)
device = 'cpu' #torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Download and extract dataset
!wget https://download.pytorch.org/tutorial/hymenoptera_data.zip
!unzip -q hymenoptera_data.zip
```

```python
# Data augmentation and normalization for training
# Just normalization for validation
train_transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
val_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

data_dir = './hymenoptera_data'
train_dataset = ImageFolder(os.path.join(data_dir, 'train'), train_transform)
val_dataset = ImageFolder(os.path.join(data_dir, 'val'), val_transform)
train_dataloader = DataLoader(train_dataset, batch_size=4,shuffle=True, num_workers=2)
val_dataloader = DataLoader(val_dataset, batch_size=4,shuffle=True, num_workers=2)
class_names = train_dataset.classes
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(train_dataloader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
def train_model(model, train_dataloader, criterion, optimizer, num_epochs=25):

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print('-' * 10)
        model.train()
        running_loss = 0.0
        running_corrects = 0

        # Iterate over data.
        for i, (inputs, labels) in enumerate(train_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward
            # track history if only in train
            with torch.set_grad_enabled(True):
              outputs = model(inputs)
              _, preds = torch.max(outputs, 1)
              loss = criterion(outputs, labels)

              # backward + optimize only if in training phase
              loss.backward()
              optimizer.step()

            # statistics
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_dataloader.dataset)
        epoch_acc = running_corrects.double() / len(train_dataloader.dataset)

        print(f'Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        print()
```

```python
def visualize_model(model, num_images=10):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(val_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'predicted: {class_names[preds[j]]}')
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
from torchvision import models

model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
model_ft.fc = nn.Linear(num_ftrs, 2)

model_ft = model_ft.to(device)

loss_func = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 5
train_model(model_ft, train_dataloader, loss_func, optimizer, num_epochs)
```

```python
visualize_model(model_ft)
```

**1.Answer:**
--

Data augmentation is a critical technique in training deep learning models, particularly for tasks like image classification, object detection, and segmentation. Here's why augmentations like resize, crop, flip, rotation, etc., are important:

**Increased Dataset Size:** Data augmentation artificially increases the size of the training dataset by generating new, slightly modified versions of the existing data. This helps prevent overfitting by exposing the model to a more diverse set of examples.

**Improved Generalization:** By presenting the model with various versions of the same data (e.g., different crops, rotations, or flips), it learns to focus on the essential features of the objects rather than on irrelevant details specific to the training examples.

**Robustness to Variability:** Augmentations simulate real-world variability that the model might encounter during inference. For example, resizing and cropping handle variations in image sizes, rotations account for object orientations, and flips address horizontal or vertical symmetry.

**Regularization:** Augmentations act as a form of regularization, which helps prevent overfitting. When a model is exposed to diverse versions of the data during training, it becomes less sensitive to small variations in the input and hence more robust.

**Reduced Dependency on Annotated Data:** Data augmentation can help alleviate the need for large annotated datasets. By generating synthetic examples, it reduces the reliance on collecting vast amounts of labeled data, which can be time-consuming and costly.

**Domain Adaptation:** Augmentations can help in domain adaptation tasks by making the model more robust to variations in the target domain. For instance, if the testing environment has different lighting conditions, augmentations like brightness adjustment can help the model generalize better.

**Enhanced Performance:** Properly chosen augmentations can improve the model's performance on the task at hand. For example, specific augmentations might be particularly helpful in certain scenarios, like affine transformations for geometric object detection tasks.

**Data Balance:** Augmentations can help address class imbalance issues by generating synthetic examples for underrepresented classes, thus improving the model's ability to learn from all classes equally.

Overall, by using augmentations like resize, crop, flip, etc., practitioners can improve the robustness, generalization, and performance of their deep learning models, leading to better results in various tasks.

**2.Answer:**
--
**Lower Accuracy:** The model trained from scratch without pretrained weights would likely achieve lower accuracy compared to the model trained with pretrained weights. This is because the pretrained weights already contain useful features learned from a vast amount of data, while the scratch-trained model would have to learn these features during training.

**Slower Convergence:** Training from scratch can take longer to converge compared to fine-tuning pretrained models. Without the advantage of pretrained weights, the model starts with random initializations, which might require more iterations to learn useful representations.

**Overfitting:** There's a higher risk of overfitting when training from scratch, especially with limited data. The model might have a tendency to memorize the training examples rather than learn generalizable features.

**Increased Training Time:** Training from scratch can be computationally more expensive, as the model needs to learn all the features from the data. This requires more time and resources compared to fine-tuning pretrained models.

**3.Answer:**
--
**Transferability of Features:** Pretrained models, especially those trained on large and diverse datasets like ImageNet, learn generic features that are often transferable across different tasks and domains. Lower-level features like edges, textures, and basic shapes are likely to be useful for various image recognition tasks, including ants vs bees classification.

**Task Similarity:** If the tasks are similar in terms of visual characteristics (e.g., both involve natural images with animals), the pretrained features might be highly relevant and beneficial. In this case, the pretrained model would have already learned representations that are useful for distinguishing between different objects, despite differences in the specific dataset.

**Fine-Tuning:** Even if the pretrained model was trained on a different dataset, fine-tuning the model on the ants vs bees dataset can adapt the learned features to better suit the new task. By updating the weights during fine-tuning, the model can adjust its learned representations to be more aligned with the target domain.

However, there are scenarios where using a pretrained model from a significantly different dataset might not yield good performance:

**Domain Shift:** If the domain shift between the pretrained dataset and the target dataset is too significant (e.g., the pretrained model was trained on medical images, and the target task involves natural images of insects), the pretrained features may not be as relevant or useful.

**Task Complexity:** If the tasks are vastly different in terms of complexity, the pretrained features might not be sufficient. For example, if the pretrained model was trained for fine-grained categorization of bird species, it might not generalize well to distinguishing between ants and bees due to the difference in the level of detail required for classification.

**Limited Data:** If the ants vs bees dataset is small, using a pretrained model from a significantly different dataset might lead to overfitting, as the model may rely too heavily on the pretrained features and fail to adapt well to the new data.

In summary, while using a pretrained model can often provide a good starting point for transfer learning, the success depends on the similarity between the pretrained dataset and the target dataset, as well as the complexity of the task and the amount of available data.
