# Data preparation

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the iris dataset
iris = load_iris()
X = iris['data']
y = iris['target']
names = iris['target_names']
feature_names = iris['feature_names']

# Scale data to have mean 0 and variance 1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data set into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=2)
```

# Visualising dataset

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
for target, target_name in enumerate(names):
    X_plot = X[y == target]
    ax1.plot(X_plot[:, 0], X_plot[:, 1],
             linestyle='none',
             marker='o',
             label=target_name)
ax1.set_xlabel(feature_names[0])
ax1.set_ylabel(feature_names[1])
ax1.axis('equal')
ax1.legend();

for target, target_name in enumerate(names):
    X_plot = X[y == target]
    ax2.plot(X_plot[:, 2], X_plot[:, 3],
             linestyle='none',
             marker='o',
             label=target_name)
ax2.set_xlabel(feature_names[2])
ax2.set_ylabel(feature_names[3])
ax2.axis('equal')
ax2.legend();
```

# MLP for multiclass classification

```python
import torch
import torch.nn.functional as F
import torch.nn as nn
from torch.autograd import Variable
import tqdm
```

```python
# Defining the model architecture
class Model(nn.Module):
    def __init__(self, input_dim):
        super(Model, self).__init__()
        self.layer1 = nn.Linear(input_dim, 50)
        self.layer2 = nn.Linear(50, 50)
        self.layer3 = nn.Linear(50, 3)

    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        x = F.softmax(self.layer3(x), dim=1)
        return x
```

```python
# Instantiating the model, using Adam optimiser, and Cross Entropy Loss, which is quite commonlu used for classification tasks.
model     = Model(X_train.shape[1])
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn   = nn.CrossEntropyLoss()
```

# Train the model

```python
# Train for 100 epochs
EPOCHS  = 100
X_train = torch.from_numpy(X_train).float()
X_test= torch.from_numpy(X_test).float()
y_test = torch.from_numpy(y_test)
y_train = torch.from_numpy(y_train)


loss_list     = np.zeros((EPOCHS,))
accuracy_list = np.zeros((EPOCHS,))

for epoch in tqdm.trange(EPOCHS):
    y_pred = model(X_train)
    loss = loss_fn(y_pred, y_train)
    loss_list[epoch] = loss.item()

    # Zero gradients
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    with torch.no_grad():
        y_pred = model(X_test)
        correct = (torch.argmax(y_pred, dim=1) == y_test).type(torch.FloatTensor)
        accuracy_list[epoch] = correct.mean()
```

# Plot training progress

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(12, 6), sharex=True)

ax1.plot(accuracy_list)
ax1.set_ylabel("validation accuracy")
ax2.plot(loss_list)
ax2.set_ylabel("validation loss")
ax2.set_xlabel("epochs");
```

Changing the number of hidden layers:

```python
class Model2(nn.Module):
    def __init__(self, input_dim):
        super(Model2, self).__init__()
        self.layer1 = nn.Linear(input_dim, 50)
        self.layer2 = nn.Linear(50, 50)
        self.layer3 = nn.Linear(50, 50)  # Additional hidden layer
        self.layer4 = nn.Linear(50, 3)

    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        x = F.relu(self.layer3(x))  # Additional hidden layer
        x = F.softmax(self.layer4(x), dim=1)
        return x

model2 = Model2(X_train.shape[1])
```

Changing the number of neurons in the hidden layer

```python
class Model3(nn.Module):
    def __init__(self, input_dim):
        super(Model3, self).__init__()
        self.layer1 = nn.Linear(input_dim, 100)  # Increased neurons
        self.layer2 = nn.Linear(100, 100)        # Increased neurons
        self.layer3 = nn.Linear(100, 3)

    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        x = F.softmax(self.layer3(x), dim=1)
        return x

model3 = Model3(X_train.shape[1])
```

Using a different activation function

```python
class Model4(nn.Module):
    def __init__(self, input_dim):
        super(Model4, self).__init__()
        self.layer1 = nn.Linear(input_dim, 50)
        self.layer2 = nn.Linear(50, 50)
        self.layer3 = nn.Linear(50, 3)

    def forward(self, x):
        x = F.leaky_relu(self.layer1(x))   # Leaky ReLU activation
        x = F.leaky_relu(self.layer2(x))   # Leaky ReLU activation
        x = F.softmax(self.layer3(x), dim=1)
        return x

model4 = Model4(X_train.shape[1])
```

code to train and compare the models:

```python
# Function to train a model and return training loss and validation accuracy
def train_model(model, optimizer, loss_fn, X_train, y_train, X_test, y_test, epochs=100):
    loss_list = np.zeros((epochs,))
    accuracy_list = np.zeros((epochs,))

    for epoch in tqdm.trange(epochs):
        # Forward pass
        y_pred = model(X_train)
        loss = loss_fn(y_pred, y_train)
        loss_list[epoch] = loss.item()

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Validation
        with torch.no_grad():
            y_pred = model(X_test)
            correct = (torch.argmax(y_pred, dim=1) == y_test).type(torch.FloatTensor)
            accuracy_list[epoch] = correct.mean()

    return loss_list, accuracy_list

# Train each model
model2 = Model2(X_train.shape[1])
optimizer2 = torch.optim.Adam(model2.parameters(), lr=0.001)
loss_fn2 = nn.CrossEntropyLoss()
loss_list2, accuracy_list2 = train_model(model2, optimizer2, loss_fn2, X_train, y_train, X_test, y_test)

model3 = Model3(X_train.shape[1])
optimizer3 = torch.optim.Adam(model3.parameters(), lr=0.001)
loss_fn3 = nn.CrossEntropyLoss()
loss_list3, accuracy_list3 = train_model(model3, optimizer3, loss_fn3, X_train, y_train, X_test, y_test)

model4 = Model4(X_train.shape[1])
optimizer4 = torch.optim.Adam(model4.parameters(), lr=0.001)
loss_fn4 = nn.CrossEntropyLoss()
loss_list4, accuracy_list4 = train_model(model4, optimizer4, loss_fn4, X_train, y_train, X_test, y_test)

# Plotting the results
fig, (ax1, ax2) = plt.subplots(2, figsize=(12, 10), sharex=True)

ax1.plot(loss_list2, label='2 Hidden Layers')
ax1.plot(loss_list3, label='Increased Neurons')
ax1.plot(loss_list4, label='Leaky ReLU')
ax1.set_ylabel("Validation Loss")
ax1.legend()

ax2.plot(accuracy_list2, label='2 Hidden Layers')
ax2.plot(accuracy_list3, label='Increased Neurons')
ax2.plot(accuracy_list4, label='Leaky ReLU')
ax2.set_ylabel("Validation Accuracy")
ax2.set_xlabel("Epochs")
ax2.legend()

plt.show()
```

Model 3, with increased neurons in hidden layers, outperforms other models with lower validation loss and higher accuracy, indicating its superior capacity to capture complex patterns. Model 4, utilizing Leaky ReLU activation, shows smoother convergence and slightly improved accuracy compared to the original ReLU, suggesting its effectiveness in mitigating the vanishing gradient problem. However, adding an extra hidden layer (Model 2) doesn't notably enhance performance, suggesting that increasing neuron capacity or using alternative activations yield more significant improvements.
